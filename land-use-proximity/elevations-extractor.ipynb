```python
import requests

url = "https://api.open-elevation.com/api/v1/lookup"

coords = df[["latitude", "longitude"]].values.tolist()

# split coords into chunks of 50
num_chunks = len(coords) // 50
chunks = [coords[i:i + 50] for i in range(0, len(coords), 50)]

results = []

for coords in chunks:

    lat_long_pairs = [{ "latitude": coord[0], "longitude": coord[1] } for coord in coords]
    body = {
        "locations": lat_long_pairs
    }

    resp = requests.post(url, json = body)
    resp_json = json.loads(resp.content.decode("utf-8"))
    results.append(resp_json["results"])
```

```python
import pandas as pd
elev_file = None
with open("elevation.json", "r") as f:
    elev_file = f.read()
data = json.loads(elev_file)[0]
lats = [coord["latitude"] for coord in data]
lons = [coord["longitude"] for coord in data]
elevs = [coord["elevation"] for coord in data]
df = pd.DataFrame({"latitude": lats, "longitude": lons, "elevation": elevs})
df
```

```python
# save to csv
df.to_csv("elevations.csv", index=False)
```
